```python
import matplotlib.pyplot as plt
import pandas as pd

import matplotlib.animation as animation
%matplotlib inline
%matplotlib nbagg
```

```python
path =  [1,2,3,4,5,6,7,1]
coordinate =  {1:(1,2), 2:(2,4),3:(14,4),4:(2,1),5:(19,23),6:(12,1),7:(34,6)}
x, y = zip(* [coordinate[i] for i in path])
```

```python
def draw (path,  coordinate, save = False):
    %matplotlib inline
    %matplotlib nbagg
    fig =  plt.figure()
    x, y = zip(* [coordinate[i] for i in path])
    ims_list =  []
    plt.scatter(x[0],y[0],color =  'r')
    plt.scatter(x[1:-1],y[1:-1],color =  'g')
    for i in range(1,len(path)+1):
        im = plt.plot(x[:i],y[:i],'b')
        ims_list.append(im)
    ani = animation.ArtistAnimation(fig, ims_list, interval=150, repeat = False)
    return ani
```

```python
draw(path, coordinate, save=True)
```

```python
fig =  plt.figure()
x, y = zip(* [coordinate[i] for i in path])
ims_list =  []
plt.scatter(x[0],y[0],color =  'r')
plt.scatter(x[1:-1],y[1:-1],color =  'g')
for i in range(len(path)+1):
    im = plt.plot(x[:i],y[:i],'b')
    ims_list.append(im)
ani = animation.ArtistAnimation(fig, ims_list, interval=150, repeat = False)
plt.show()
```
