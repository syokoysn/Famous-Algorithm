### 交差しているかを判定する関数

```python
def cross_check(a,b,c,d):
    ax,ay = a[0],a[1]; bx,by = b[0],b[1]; cx,cy = c[0],c[1]; dx,dy = d[0],d[1]
    ta = (cx - dx) * (ay - cy) + (cy - dy) * (cx - ax)
    tb = (cx - dx) * (by - cy) + (cy - dy) * (cx - bx)
    tc = (ax - bx) * (cy - ay) + (ay - by) * (ax - cx)
    td = (ax - bx) * (dy - ay) + (ay - by) * (ax - dx)
    return (tc * td < 0) &( ta * tb < 0)

def two_opt(path, coordinate, n =  0):
    if sys.getrecursionlimit()-1 ==  n :
        return path 
    edges =  [(path[i], path[i+1])for i in range(len(path )-1)]
    for i in range(len(edges)):
        for j in range(i+2,len(edges)):
            a,b = edges[i][0],edges[i][1]; c, d =  edges[j][0], edges[j][1]
            if cross_check(coordinate[a],coordinate[b],coordinate[c], coordinate[d]):
                path = path[:i+1]+ path[i+1:j+1][::-1]+path[j+1:]
                return two_opt(path, coordinate, n+1)
    return path 
```

## 非再帰

```python
def two_opt2(path, coordinate):
    def cross_check(a,b,c,d):
        ax,ay = a[0],a[1]; bx,by = b[0],b[1]; cx,cy = c[0],c[1]; dx,dy = d[0],d[1]
        ta = (cx - dx) * (ay - cy) + (cy - dy) * (cx - ax)
        tb = (cx - dx) * (by - cy) + (cy - dy) * (cx - bx)
        tc = (ax - bx) * (cy - ay) + (ay - by) * (ax - cx)
        td = (ax - bx) * (dy - ay) + (ay - by) * (ax - dx)
        return (tc * td < 0) &( ta * tb < 0)
    while True :
        edges =  [(path[i], path[i+1])for i in range(len(path )-1)]
        for i in range(len(edges)):
            #print(i)
            for j in range(i+2,len(edges)):
                a,b = edges[i][0],edges[i][1]; c, d =  edges[j][0], edges[j][1]
                if cross_check(coordinate[a],coordinate[b],coordinate[c], coordinate[d]):
                    path = path[:i+1]+ path[i+1:j+1][::-1]+path[j+1:]
                    break
                else :
                    continue
            else :
                continue
            break
        else :
            break
    return path 
```

```python
new_path =  two_opt2(path2, coordinate2)
x,y =  zip(*[coordinate2[i] for i in new_path])
plt.scatter(x,y)
plt.plot(x,y,'-')
```

### 非再帰の改良

```python
process =  []
def two_opt3(path, coordinate):
    process.append(path)
    def cross_check(a,b,c,d):
        ax,ay = a[0],a[1]; bx,by = b[0],b[1]; cx,cy = c[0],c[1]; dx,dy = d[0],d[1]
        ta = (cx - dx) * (ay - cy) + (cy - dy) * (cx - ax)
        tb = (cx - dx) * (by - cy) + (cy - dy) * (cx - bx)
        tc = (ax - bx) * (cy - ay) + (ay - by) * (ax - cx)
        td = (ax - bx) * (dy - ay) + (ay - by) * (ax - dx)
        return (tc * td < 0) &( ta * tb < 0)
    n = 0
    while True :
        edges =  [(path[i], path[i+1])for i in range(len(path )-1)]
        for i in range(n,len(edges)):
            #print(i)
            for j in range(i+2,len(edges)):
                a,b = edges[i][0],edges[i][1]; c, d =  edges[j][0], edges[j][1]
                #print(a,b, c,d)
                if cross_check(coordinate[a],coordinate[b],coordinate[c], coordinate[d]):
                    #print('cross')
                    path = path[:i+1]+ path[i+1:j+1][::-1]+path[j+1:]
                    # n の変更
                    process.append(path)
                    break
                else :
                    continue
            else :
                continue
            break
        else :
            break
    return path 
```

```python
new_path =  two_opt3(path2, coordinate2)
```

```python
for i in process :
    x,y =  zip(*[coordinate2[i] for i in i])
    plt.scatter(x,y)
    plt.plot(x,y,'-')
```

```python
%matplotlib nbagg
fig =  plt.figure(figsize =(10,10))
x, y = zip(* [coordinate2[i] for i in process[0]])
ims_list =  []
plt.scatter(x[0],y[0],color =  'r')
plt.scatter(x[1:-1],y[1:-1],color =  'g')
for p in process:
    x, y = zip(* [coordinate2[i] for i in p])
    im = plt.plot(x,y,'b')
    ims_list.append(im)
ani = animation.ArtistAnimation(fig, ims_list, interval=400, repeat = False)
plt.show()
#ani.save('test.gif')
```
